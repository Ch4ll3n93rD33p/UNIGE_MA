```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import random
from copy import deepcopy
```

We start by defining useful functions for image and histogram display.

```python
def display_img(images, labels, rows, cols, cmap = None) :
    ''' function that allows us to display one or more images '''

    axes = []
    fig = plt.figure()

    for i in range(rows * cols) :
        axes.append(fig.add_subplot(rows, cols, i + 1))
        plt.imshow(images[i], cmap)
        plt.title(labels[i])
        plt.axis('off')
    fig.tight_layout()
    plt.show()

def display_histo(histograms, tabs, labels, rows, cols) :
    ''' function that allows us to display one or more histogram '''

    _, axs = plt.subplots(rows, cols, figsize=(20,5))

    for i in range(rows * cols) :
        axs[i].plot(histograms[i], tabs[i])
        axs[i].set_title(labels[i])
    plt.show()
```

```python
def MSE(X, Y) :
    ''' function that computes the Mean Squared Error between X and Y '''
    if X.shape != Y.shape :
        return 0 
    else:
        return np.sum([np.sum([(X[i, j]- Y[i, j])**2 for j in range(X.shape[1])]) for i in range(X.shape[0])]) / (X.shape[0]*X.shape[1])
    
def PSNR(X, Y) :
    ''' function that computes the Peak Signal to Noise Ratio between X and Y '''

    mse = MSE(X,Y)
    if mse == 0:
        return 0
    else:
        return 10 * np.log10(255**2/mse)
```

## 1 Watermark Embedding and Channel modelling


### Exercise 1.1

```python
def create_watermark(w, N, H, W) :
    ''' function that creates the watermark '''
    
    watermark = np.int_(np.zeros(w.shape))
    d = []
    while N > 0:
        r = np.random.randint(0, w.shape[0])
        if r not in d:
            watermark[r] = w[r]
            N = N - 1
            d.append(r)
    watermark = watermark.reshape(H, W)

    return watermark
```

```python
img_x = cv2.imread('liftingbody.png',0)
H,W = img_x.shape

m = np.int_(np.floor(np.random.uniform(0, 1, H*W)*2))
w = np.where(m == 0, -1, m)

wm_len = int((H*W)*0.5)
watermark = create_watermark(w, wm_len, H, W)

# watermarked image
img_y = np.clip(img_x+watermark, 0, 255)

# AGWN
img_z = np.asarray([[np.random.normal(0,1) for i in range(W)] for j in range(H)])

# watermarked image + AGWN
img_v = np.clip(img_y+img_z, 0, 255)
```

```python
images = [img_x, img_y, img_v]
labels = ["original image", "watermarked image", "attacked watermarked image"]

display_img(images, labels, 1, 3, cmap = "gray")

print("PSNR(x, y) = ", PSNR(img_x, img_y))
print("PSNR(y, v) = ", PSNR(img_y, img_v))
print("PSNR(x, v) = ", PSNR(img_x, img_v))
```

We started by watermarking the image, then added the AGWN.  
Visually, there is no difference between the watermarked image and the original image, but since the PSNR is not zero, we have confirmation that the two images are different.  
The difference with the attacked image is also hardly noticeable. It reduced the watermark a bit, but was not very successful.

## 2 Non-blind watermark detection


```python
def linear_correlation(x,y):
    H,W = x.shape
    s = 0
    N = 0
    for i in range(H):
        for j in range(W):
            if x[i,j] != 0:
                s += x[i,j]*y[i,j]
                N += 1
    return s/N
```

### Exercise 2.1

```python
non_blind_wm = img_v-img_x
rho_non_blind = linear_correlation(watermark, non_blind_wm)

print(rho_non_blind)
```

Here we just take the difference between the original image and the attacked watermarked image to try to recreate the watermark. As we can see we have realy good results.

## 3 Blind watermark detection using the Maximum Likelihood estimate

```python
def convolution(x,N):
    res = deepcopy(x)
    H,W = x.shape
    m = int((N-1)/2)
    for i in range(m,H-m):
        for j in range(m,W-m):
            res[i,j] = np.int64(np.mean(x[i-m:i+m+1, j-m:j+m+1]))
    return res
```

### Exercise 3.1


```python
local_mean_v = convolution(img_v,3)
blind_wm = img_v-local_mean_v

print("PSNR(x, local_mean_v) = ", PSNR(img_x, local_mean_v))
```

### Exercise 3.2

```python
rho_blind = linear_correlation(watermark, blind_wm)

print(rho_blind)
```

Without knowledge of the original image (blind detection of the watermark), the result is a bit worse, but the linear correlation with the original watermark is still very close to 1. We can conclude that the blind detection is also very efficient.
